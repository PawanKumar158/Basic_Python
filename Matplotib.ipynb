Importing python libraries to use this learning Matplot Library.

```python
import numpy as np
from matplotlib import pyplot as plt
```

# Matplotlib is a comprehensive library for creating static, animated, and interactive *visualizations* in *Python*.

It is widely used for data visualization and is built on top of NumPy arrays. Matplotlib makes easy things easy and hard things possible. You can create publication quality plots, make interactive figures that can zoom, pan, update, customize visual style and layout, export to many file formats, embed in JupyterLab and Graphical User Interfaces, and use a rich array of third-party packages built on Matplotlib.


Arranging data to perform visualizations some operations

```python
x = np.arange(1,21)
x
```

```python
y=x*2
y
```

Line Plot showing between x and y data sets

```python
plt.plot(x,y)
plt.show()
```

Title and data labels we can add on this line plots.

```python
plt.plot(x,y)
plt.title('Time and Work Line Plot')
plt.xlabel('Work')
plt.ylabel('Time')
plt.show()
```

Its Default Color is blue and line style is solid line but we can change both things easily.

```python
plt.plot(x,y ,color='g', linestyle=":", linewidth=3)  # Dotted greenline
plt.show()
```

# Subploting


```python
plt.subplot(1,2,1)
plt.plot(x,y ,color='g', linestyle=":", linewidth=3)  # Dotted greenline
plt.subplot(1,2,2)
plt.plot(x,y ,color='red', linestyle="-", linewidth=3)  # Dotted greenline
plt.grid(True)        # Grid
plt.show()
```

# Bar Plot

```python
student = { 'Bob': 87, "Mak":65, 'Zoks':24,'Boot': 37, "Mac":62, 'Zoo':44}
name= list(student.keys())  # key value of student dictionary convert into list name
Value = list(student.values()) #value of student dictionary convert into value list
plt.bar(name,Value)
plt.show()
```

Title and labels

```python
plt.bar(name,Value)
plt.title('Student Mark Sheet')
plt.xlabel('Student name')
plt.ylabel('Student marks')
plt.grid(True)
plt.show()
```

subplot, and color

```python
plt.subplot(1,2,1)
plt.bar(name,Value, color="red")
plt.grid(True)
plt.subplot(1,2,2)
plt.bar(name,Value, color="orange")
plt.grid(True)
plt.show()
```

Horizontal Bar plots

```python
plt.subplot(1,2,1)
plt.barh(name,Value, color="red")
plt.grid(True)
plt.subplot(1,2,2)
plt.barh(name,Value, color="orange")
plt.grid(True)
plt.show()
```
